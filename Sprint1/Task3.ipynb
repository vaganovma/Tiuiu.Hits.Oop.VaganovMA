```
using System;
using System.Collections.Generic;

// Класс Товар (Product)
public class Product
{
    // Приватные поля 
    private string name;
    private decimal price;
    private int quantity;
    
    // Защищенное поле 
    protected string category;
    
    // Внутреннее поле 
    internal string description;
    
    // Конструктор класса для инициализации полей
    public Product(string name, decimal price, int quantity, string category, string description)
    {
        this.name = name;
        this.price = price;
        this.quantity = quantity;
        this.category = category;
        this.description = description;
    }
    
    // Публичные свойства (геттеры) для доступа к приватным полям
    public string Name => name;
    public decimal Price => price;
    public int Quantity => quantity;
    
    // Метод для получения общей стоимости всех единиц товара
    public decimal GetTotalPrice()
    {
        return price * quantity;
    }
    
    // Метод для обновления количества товара
    public void UpdateQuantity(int amount)
    {
        quantity += amount;
        // Защита от отрицательного количества
        if (quantity < 0) quantity = 0;
    }
    
    // Метод для получения полной информации о товаре
    public string GetProductInfo()
    {
        return $"Товар: {name}, Цена: {price:C}, Количество: {quantity}, " +
               $"Категория: {category}, Описание: {description}, " +
               $"Общая стоимость: {GetTotalPrice():C}";
    }
}

// Класс Продавец (Seller)
public class Seller
{
    // Приватные поля
    private string name;
    private string employeeId;
    
    // Защищенное поле
    protected decimal salary;
    
    // Внутреннее поле
    internal string contactInfo;
    
    // Приватный список товаров продавца
    private List<Product> products = new List<Product>();
    
    // Конструктор класса
    public Seller(string name, string employeeId, decimal salary, string contactInfo)
    {
        this.name = name;
        this.employeeId = employeeId;
        this.salary = salary;
        this.contactInfo = contactInfo;
    }
    
    // Публичные свойства для доступа к приватным полям
    public string Name => name;
    public string EmployeeId => employeeId;
    
    // Метод для добавления товара в список продавца
    public void AddProduct(Product product)
    {
        products.Add(product);
        Console.WriteLine($"Товар '{product.Name}' добавлен продавцу {name}");
    }
    
    // Метод для продажи товара (уменьшение количества)
    public void SellProduct(Product product, int quantity)
    {
        if (product.Quantity >= quantity)
        {
            product.UpdateQuantity(-quantity);
            Console.WriteLine($"Продано {quantity} единиц товара '{product.Name}'");
        }
        else
        {
            Console.WriteLine($"Недостаточно товара '{product.Name}' для продажи");
        }
    }
    
    // Метод для получения информации о продавце
    public string GetSellerInfo()
    {
        return $"Продавец: {name}, ID: {employeeId}, " +
               $"Зарплата: {salary:C}, Контакты: {contactInfo}, " +
               $"Товаров в управлении: {products.Count}";
    }
    
    // Метод для получения списка товаров продавца
    public List<Product> GetProducts()
    {
        return products;
    }
}

// Класс Магазин (Store)
public class Store
{
    // Приватные поля
    private string storeName;
    private string location;
    
    // Публичное поле
    public string storeHours;
    
    // Приватный список продавцов магазина
    private List<Seller> sellers = new List<Seller>();
    
    // Конструктор класса
    public Store(string storeName, string location, string storeHours)
{
        this.storeName = storeName;
        this.location = location;
        this.storeHours = storeHours;
    }
    
    // Публичные свойства для доступа к приватным полям
    public string StoreName => storeName;
    public string Location => location;
    
    // Метод для добавления продавца в магазин
    public void AddSeller(Seller seller)
    {
        sellers.Add(seller);
        Console.WriteLine($"Продавец {seller.Name} добавлен в магазин {storeName}");
    }
    
    // Метод для перечисления всех товаров в магазине
    public void ListProducts()
    {
        Console.WriteLine($"\nВсе товары в магазине '{storeName}':");
        
        foreach (var seller in sellers)
        {
            foreach (var product in seller.GetProducts())
            {
                Console.WriteLine($"- {product.GetProductInfo()}");
            }
        }
    }
    
    // Метод для получения информации о магазине
    public string GetStoreInfo()
    {
        return $"Магазин: {storeName}, Адрес: {location}, " +
               $"Часы работы: {storeHours}, Продавцов: {sellers.Count}";
    }
    
    // Метод для получения списка продавцов
    public List<Seller> GetSellers()
    {
        return sellers;
    }
}

// Основной класс программы
class Program
{
    static void Main(string[] args)
    {
        Console.WriteLine("=== СПРИНТ РЕВЬЮ #1 ===\n");
        
        // Создание экземпляра товара
        Product laptop = new Product(
            name: "Ноутбук HP Pavilion",
            price: 45000m,
            quantity: 10,
            category: "Электроника",
            description: "Игровой ноутбук с процессором Intel i7"
        );
        
        // Создание экземпляра продавца
        Seller seller1 = new Seller(
            name: "Иван Петров",
            employeeId: "EMP001",
            salary: 35000m,
            contactInfo: "ivan@store.com"
        );
        
        // Создание экземпляра магазина
        Store electronicsStore = new Store(
            storeName: "ЭлектроникСтар",
            location: "ул. Пушкина, 25",
            storeHours: "9:00-21:00"
        );
        
        // Добавление продавца в магазин
        electronicsStore.AddSeller(seller1);
        
        // Добавление товара продавцу
        seller1.AddProduct(laptop);
        
        // Вывод информации о товаре
        Console.WriteLine($"\nИнформация о товаре: {laptop.GetProductInfo()}");
        
        // Продажа товара
        seller1.SellProduct(laptop, 3);
        
        // Проверка количества после продажи
        Console.WriteLine($"Количество после продажи: {laptop.Quantity}");
        
        // Вывод информации о продавце
        Console.WriteLine($"\nИнформация о продавце: {seller1.GetSellerInfo()}");
        
        // Вывод информации о магазине
        Console.WriteLine($"\nИнформация о магазине: {electronicsStore.GetStoreInfo()}");
        
        // Перечисление всех товаров в магазине
        electronicsStore.ListProducts();
        
        // Демонстрация работы с разными модификаторами доступа
        Console.WriteLine("\n=== ДЕМОНСТРАЦИЯ МОДИФИКАТОРОВ ДОСТУПА ===");
        
        // Доступ к public свойствам
        Console.WriteLine($"Public доступ: {laptop.Name}, {laptop.Price}");
        
        // Доступ к internal полю (в пределах сборки)
        Console.WriteLine($"Internal доступ: {laptop.description}");
        
        // Попытка доступа к private полям приведет к ошибке компиляции:
        // Console.WriteLine(laptop.name); // Ошибка!
        // Console.WriteLine(seller1.name); // Ошибка!
        
        // Попытка доступа к protected полям также приведет к ошибке:
        // Console.WriteLine(laptop.category); // Ошибка!
        // Console.WriteLine(seller1.salary); // Ошибка!

    }
}
```
